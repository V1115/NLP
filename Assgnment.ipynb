```python
import pandas as pd
import numpy as np

 #MODEL data for section A
customers_data = {
    'Customer_ID': [1, 2, 3, 4, 5, 6],
    'Age': [25, 150, 34, -1, 0, np.nan],
    'Purchase_Date': ['2024-05-12', '12/06/24', 'June 15, 2024', 'invalid_date', '2024-01-01', '03-03-2024'],
    'Country': ['India', 'india', 'INDIA ', 'Ind', ' India', 'IND']
}
df_customers = pd.DataFrame(customers_data)


 # MODEL data for section B
orders_data = {
    'Order_ID': [101, 102, 103, 104],
    'Customer_ID': [1, 2, 3, 4],
    'Product_ID': ['P1', 'P2', 'P1', 'P99'],
    'Quantity': [2, 1, 5, 3]
}
DF_Orders = pd.DataFrame(orders_data)

product_data = {
    'Product_ID': ['P1', 'P2', 'P3'],
    'Product_Name': ['Laptop', 'Mouse', 'Keyboard'],
    'Unit_Price': [1000, 25, 50],
    'Category': ['Electronics', 'Accessories', 'Accessories']
}
DF_ProductDetails = pd.DataFrame(product_data)

print("--- ORIGINAL CUSTOMER DATA ---")
print(df_customers, "\n")



```

## SECTION A: DATA CLEANING

### 1. Handling Invalid Ages

```python
invalid_ages = df_customers[(df_customers['Age'] <= 0) | (df_customers['Age'] > 100)]

df_customers.loc[(df_customers['Age'] <= 0) | (df_customers['Age'] > 120), 'Age'] = np.nan
median_age = df_customers['Age'].median()
df_customers['Age'] = df_customers['Age'].fillna(median_age)
```

### 2. Text Normalization (Country)

```python
df_customers['Country'] = (df_customers['Country']
                           .str.strip()
                           .str.title()
                           .replace({'Ind': 'India'})
                          )
```

### 3. Date Standardization (Purchase_Date)

```python
df_customers['Purchase_Date'] = pd.to_datetime(df_customers['Purchase_Date'], errors='coerce')

print(" CLEANED CUSTOMER DATA (SECTION A) ")
print(df_customers, "\n")
```

## SECTION B: DATA WRANGLING & TRANSFORMATION

### 1. Merging DataFrames and Handling Missing Product Details

```python
merged_df = pd.merge(DF_Orders, DF_ProductDetails, on='Product_ID', how='left')

merged_df['Unit_Price'] = merged_df['Unit_Price'].fillna(0)
merged_df['Category'] = merged_df['Category'].fillna('Unknown Category')
merged_df['Product_Name'] = merged_df['Product_Name'].fillna('Unknown Product')

print("--- MERGED DATAFRAME (SECTION B) ---")
print(merged_df, "\n")
```

### 2. Aggregation and Grouping (Total Sales per Category)

```python
merged_df['Total_Sales'] = merged_df['Quantity'] * merged_df['Unit_Price']

sales_per_category = merged_df.groupby('Category')['Total_Sales'].sum().reset_index()

print("--- TOTAL SALES PER CATEGORY ---")
print(sales_per_category)
```

3 Handling Missing Values

```python
merged_df['Unit_Price'] = merged_df['Unit_Price'].fillna(0)
merged_df['Category'] = merged_df['Category'].fillna('Unknown Category')
merged_df['Product_Name'] = merged_df['Product_Name'].fillna('Unknown Product')

print("--- MERGED DATAFRAME (SECTION B) ---")
print(merged_df, "\n")
```
