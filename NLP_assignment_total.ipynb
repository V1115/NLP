## 1. Stemming vs Lemmatization

### Objective
To process raw text and compare stemming and lemmatization techniques.

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

# Input text
text = "I am running and studying better methods for learning"

# Tokenize
words = word_tokenize(text)

# Initialize tools
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Original Words:", words)

print("\n--- Stemming ---")
for word in words:
    print(word, "→", stemmer.stem(word))

print("\n--- Lemmatization ---")
for word in words:
    print(word, "→", lemmatizer.lemmatize(word, pos='v'))
```

## 3. POS Tagging and N-grams

### Objective
To perform part-of-speech tagging and use n-gram models to predict the next word in a sentence.

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

text = "I love playing football with my friends"

words = word_tokenize(text)
pos_tags = pos_tag(words)

print("POS Tags:")
for word, tag in pos_tags:
    print(word, "→", tag)
```

### Explanation

POS tagging assigns grammatical labels to each word in a sentence.  
It helps in understanding sentence structure and meaning.

Examples:
- NN → Noun
- VB → Verb
- JJ → Adjective

### N-grams and Next Word Prediction

```python
from nltk.util import ngrams

tokens = word_tokenize(text)

# Generate n-grams
unigrams = list(ngrams(tokens, 1))
bigrams = list(ngrams(tokens, 2))
trigrams = list(ngrams(tokens, 3))

print("Unigrams:\n", unigrams)
print("\nBigrams:\n", bigrams)
print("\nTrigrams:\n", trigrams)
```

```python
from collections import Counter

bigram_model = Counter(bigrams)

def predict_next(word):
    candidates = {pair[1]: count for pair, count in bigram_model.items() if pair[0] == word}
    return max(candidates, key=candidates.get) if candidates else "No prediction"

print("\nNext word after 'playing':", predict_next("playing"))
```

### Comparison

**Unigram Model:**
- Uses single words
- No context
- Least accurate

**Bigram Model:**
- Uses pairs of words
- Moderate context
- Better predictions

**Trigram Model:**
- Uses three-word sequences
- High context
- Most accurate but needs more data

### Conclusion

N-gram models improve prediction accuracy as the value of n increases. However, higher-order models require more data and computational power.

## 4. Text Cleaning

### Objective
To clean text by removing emails, hashtags, and URLs using regex and compare it with tokenization-based cleaning.

```python
import re

text = "Contact me at test@gmail.com #AI https://example.com I love NLP!"

# Regex cleaning
clean_text = re.sub(r"http\S+|www\S+|@\S+|#\S+|\S+@\S+", "", text)

print("Original Text:\n", text)
print("\nCleaned Text (Regex):\n", clean_text)
```

```python
from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)

clean_tokens = [word for word in tokens if not word.startswith('#') 
                and "@" not in word 
                and not word.startswith("http")]

clean_text_token = " ".join(clean_tokens)

print("\nCleaned Text (Tokenization):\n", clean_text_token)
```

### Comparison

**Regex Cleaning:**
- Very fast
- Removes patterns efficiently
- Not flexible for complex cases

**Tokenization Cleaning:**
- More flexible
- Can apply custom logic
- Slightly slower

### Conclusion

Regex is suitable for quick and simple cleaning tasks, while tokenization-based methods are better for more complex and customizable text processing.

## 5. Parsing (CKY vs Earley)

### Objective
To parse sentences using context-free grammar and compare CKY and Earley parsing methods.

```python
from nltk import CFG
from nltk.parse import ChartParser

# Define grammar
grammar = CFG.fromstring("""
S -> NP VP
NP -> 'John'
VP -> V NP
V -> 'eats'
NP -> 'apple'
""")

# Create parser
parser = ChartParser(grammar)

sentence = "John eats apple".split()

print("Parse Tree:\n")
for tree in parser.parse(sentence):
    print(tree)
```

### Explanation

Parsing is used to analyze the grammatical structure of a sentence.

In this example:
- S → Sentence
- NP → Noun Phrase
- VP → Verb Phrase

The parser builds a tree representing sentence structure.

### Comparison

**CKY Parser:**
- Uses dynamic programming
- Works only with Chomsky Normal Form (CNF)
- Efficient for structured grammars

**Earley Parser:**
- More flexible
- Works with any CFG
- Slightly slower but more powerful

### Conclusion

CKY parsing is efficient but restricted to CNF grammars, while Earley parsing is more flexible and suitable for general-purpose parsing.

## 6. Dependency Parsing

### Objective
To extract subject–verb–object (SVO) relationships from a sentence and compare dependency parsing with constituency parsing.

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "John eats an apple"

doc = nlp(text)

print("Token | Dependency | Head")
print("--------------------------------")

for token in doc:
    print(f"{token.text:10} | {token.dep_:10} | {token.head.text}")
```

```python
subject = ""
verb = ""
obj = ""

for token in doc:
    if token.dep_ == "nsubj":
        subject = token.text
    elif token.dep_ == "ROOT":
        verb = token.text
    elif token.dep_ == "dobj":
        obj = token.text

print("\nExtracted SVO:")
print("Subject:", subject)
print("Verb:", verb)
print("Object:", obj)
```

### Comparison

**Dependency Parsing:**
- Focuses on relationships between words
- Directly extracts SVO structure
- Easier for information extraction

**Constituency Parsing:**
- Breaks sentence into phrases
- Focuses on structure (NP, VP)
- More complex to interpret

### Conclusion

Dependency parsing is more useful for extracting relationships like subject–verb–object, while constituency parsing is better for understanding hierarchical sentence structure.

## 7. CFG vs PCFG

### Objective
To generate sentences using context-free grammar and compare deterministic CFG with probabilistic CFG.

```python
from nltk import CFG
from nltk.parse.generate import generate

grammar = CFG.fromstring("""
S -> NP VP
NP -> 'John' | 'Mary'
VP -> V NP
V -> 'likes' | 'hates'
""")

print("Generated Sentences (CFG):\n")

for sentence in generate(grammar, n=5):
    print(' '.join(sentence))
```

```python
from nltk import PCFG

grammar = PCFG.fromstring("""
S -> NP VP [1.0]
NP -> 'John' [0.5] | 'Mary' [0.5]
VP -> V NP [1.0]
V -> 'likes' [0.6] | 'hates' [0.4]
""")

print("\nPCFG Grammar with Probabilities:\n")
print(grammar)
```

### Explanation

CFG generates sentences using fixed rules.

PCFG assigns probabilities to each rule, making sentence generation more realistic and data-driven.

### Comparison

**CFG (Context-Free Grammar):**
- Fixed rules
- No probabilities
- Deterministic output

**PCFG (Probabilistic CFG):**
- Rules have probabilities
- Produces more realistic language
- Handles ambiguity better

### Conclusion

PCFG is more powerful than CFG as it incorporates probabilities, making it suitable for real-world NLP tasks.

## 8. Word Sense Disambiguation (WSD)

### Objective
To determine the correct meaning of ambiguous words using dictionary-based and machine learning approaches.

```python
from nltk.wsd import lesk
from nltk.tokenize import word_tokenize

sentence = "I went to the bank to deposit money"
word = "bank"

tokens = word_tokenize(sentence)

sense = lesk(tokens, word)

print("Word:", word)
print("Context:", sentence)
print("\nPredicted Meaning:\n", sense)
```

### Explanation

Word Sense Disambiguation identifies the correct meaning of a word based on context.

Example:
- "bank" can mean a financial institution or river bank
- The algorithm selects the correct meaning based on surrounding words

### Comparison

**Dictionary-Based (Lesk):**
- Uses dictionary definitions
- Simple and fast
- Less accurate

**Machine Learning-Based:**
- Uses trained models
- More accurate
- Requires data and training

### Conclusion

Dictionary-based methods are simple but less accurate, while machine learning approaches provide better performance for word sense disambiguation.

## 9. Word Embeddings

### Objective
To represent words as numerical vectors and compare different embedding techniques in a text classification task.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love football",
    "Football is a great sport",
    "I enjoy playing games"
]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

print("Feature Names:\n", vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix:\n", X.toarray())
```

### Explanation

Word embeddings convert words into numerical vectors.

TF-IDF:
- Represents words based on frequency
- Produces sparse vectors

Word2Vec:
- Learns meaning based on context
- Produces dense vectors

### Comparison

**TF-IDF:**
- Sparse representation
- Based on frequency
- Easy to implement

**Word2Vec:**
- Dense representation
- Captures semantic meaning
- More powerful

### Conclusion

Word2Vec provides better semantic understanding of words, while TF-IDF is simpler and effective for basic tasks.

## 10. Discourse Segmentation

### Objective
To divide text into meaningful units (sentences or discourse segments) and compare rule-based and machine learning approaches.

```python
from nltk.tokenize import sent_tokenize

text = "I love football. It is a great sport. Many people enjoy it."

segments = sent_tokenize(text)

print("Discourse Segments:\n")
for i, seg in enumerate(segments, 1):
    print(f"{i}. {seg}")
```

### Explanation

Discourse segmentation divides text into smaller meaningful units.

Here, we split the paragraph into sentences using punctuation rules.

### Comparison

**Rule-Based Segmentation:**
- Uses punctuation (., !, ?)
- Simple and fast
- May fail in complex sentences

**Machine Learning-Based:**
- Uses context and patterns
- More accurate
- Requires training data

### Conclusion

Rule-based segmentation is simple and efficient, but machine learning methods provide better accuracy for complex text.

## 11. Pronoun Resolution

### Objective
To resolve pronouns in text and identify the noun they refer to.

```python
text = "John went to the store. He bought milk."

sentences = text.split(".")

# Simple heuristic: assume pronoun refers to last noun
noun = ""
pronoun_resolution = {}

for sentence in sentences:
    words = sentence.split()
    for word in words:
        if word.lower() in ["he", "she", "it", "they"]:
            pronoun_resolution[word] = noun
        else:
            noun = word

print("Pronoun Resolution:\n")
for pronoun, ref in pronoun_resolution.items():
    print(f"{pronoun} → {ref}")
```

### Explanation

Pronoun resolution identifies the noun that a pronoun refers to.

Example:
- "John went to the store. He bought milk."
- "He" refers to "John"

### Comparison

**Heuristic-Based Approach:**
- Simple rules
- Fast and easy
- Not very accurate

**Deep Learning-Based Approach:**
- Uses neural networks
- Highly accurate
- Requires large datasets

### Conclusion

Heuristic methods are simple but limited, while deep learning approaches provide better accuracy in resolving pronouns.
